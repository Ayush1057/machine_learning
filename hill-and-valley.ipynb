# **Hill and Valley Prediction**

-------------

## **Objective**

To Develop a powerful and accurate machine learning model to predict hills and valleys in a given landscape based on input data and explore and preprocess the dataset to ensure it is suitable for training and evaluating the predictive model.

## **Data Source**

The data is taken from "https://www.kaggle.com/datasets/arpitapanja/white-wine-quality"

## **Import Library**

```python
import pandas as pd 
import matplotlib.pyplot as plt
```

## **Import Data**

```python
data = pd.read_csv("/kaggle/input/white-wine-quality/Hill Valley Dataset.csv")
```

## **Describe Data**

```python
data.head()
data.info()
data.describe()
```

## **Data Preprocessing**

```python
data.isnull()
```

```python
data.dropna(inplace = True)
```

```python
data.shape
```

```python
data.columns
```

## **Define Target Variable (y) and Feature Variables (X)**

```python
y = data['Class']
x = data.drop('Class', axis = 1)
```

```python
x.shape, y.shape
```

## **Data Visualization**

```python
plt.plot(x.iloc[0,:])
plt.title("Valley")
```

```python
plt.plot(x.iloc[1,:])
plt.title("hill")
```

## **Train Test Split**

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y, random_state = 2529)
```

## **Modeling**

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression(max_iter = 5000)
```

```python
model.fit(x_train , y_train)
```

## **Prediction**

```python
y_predict = model.predict(x_test)
```

```python
y_predict
```

## **Model Evaluation**

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
print(confusion_matrix(y_test,y_predict))
```

```python
print(classification_report(y_test,y_predict))
```
